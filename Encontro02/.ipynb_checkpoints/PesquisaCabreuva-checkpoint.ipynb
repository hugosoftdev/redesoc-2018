```python
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv("cabreuva_limpa.csv", encoding='iso-8859-1')
```

```python
df_formatted = df[['RAÇA','ACREDITOEMSORTEMAISDOQUETRABALHO','ALGOMEIMPEDEDECONQUISTARASCOISAS',
                   'ACREDITONOSMEUSPLANOS', 'TANTOFAZBRANCOSOUNEGROS', 'GRUPOSSUPERIORESDEVEMCONDUZIR',
                   'PAISENCORAJADORES']]
df_formatted
```

```python
crosstabed_table = pd.crosstab(df_formatted.RAÇA, [df_formatted.ACREDITOEMSORTEMAISDOQUETRABALHO, df_formatted.GRUPOSSUPERIORESDEVEMCONDUZIR], margins=True)
```

```python
crosstabed_table
```

```python
crosstabed_table.plot()
```
